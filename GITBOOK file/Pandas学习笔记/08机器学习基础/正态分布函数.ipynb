```python
import numpy as np
import matplotlib.pyplot as plt
figure = plt.figure('标准正态分布',figsize=(8,4))
ax = figure.add_axes([0.1,0.1,0.8,0.8])
N = lambda x : (1.0/np.sqrt(np.pi*2.0))*np.exp(-x**2/2.0)
ax.plot(x,y,c = 'r',)
x =np.linspace(-5,5,100) # 100个均匀分布
y = N(x)
y_0 = N(0)
y_001= N(0.01)
y_0001= N(0.001)
ax.scatter(0,y_0,marker = '>')
ax.scatter(0.01,y_001,marker = '*',c= 'blue')
ax.scatter(0.001,y_0001,marker = '^',c='g')
plt.show()
```
